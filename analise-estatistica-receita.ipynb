```python
"""analise-estatistica-receita

analise_estatistica_inicial.ipynb
README.md

Análise Estatística de Receita Simulada

Objetivo: Avaliar a estabilidade dos gastos de uma distribuidora a partir de métricas estatísticas descritivas."""
    
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

receita = [1200, 1500, 1100, 1800, 1700, 1600, 1400, 1550, 1650, 1750]

media = np.mean(receita)
mediana = np.median(receita)
variancia = np.var(receita)
desvio_padrao = np.std(receita)
coef_var = desvio_padrao / media

print(f"Média: {media}")
print(f"Mediana: {mediana}")
print(f"Variância: {variancia}")
print(f"Desvio Padrão: {desvio_padrao}")
print(f"Coeficiente de Variação: {coef_var*100:.2f}%")

"""Interpretação
A média de gastos foi de R$ 1.525, com desvio padrão de aproximadamente R$ 219.

O coeficiente de variação foi de 14,38%, indicando variação moderada nos custos.

Isso sugere que os gastos apresentam oscilações controladas, sem indícios de alta volatilidade."""


plt.hist(receita)
plt.title("Distribuição da Receita")
plt.xlabel("Valor")
plt.ylabel("Frequência")
plt.show()
    



    
```
